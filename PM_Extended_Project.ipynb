```python
# Outlier Treatment
df[num].plot(kind='box',subplots=True, layout=(4,4),figsize=(20,18));
```

```python
outliers=["sales","capital","patents","randd","employment","tobinq","value"]
```

```python
def get_bounds (df,col):
    q1= df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    IQR= q3-q1
    upper_bound= q3+1.5*IQR
    lower_bound= q1-1.5*IQR
    return upper_bound, lower_bound
```

```python
for cols in outliers:
    upper_bound, lower_bound= get_bounds(df,cols)
    df[cols]= np.where(df[cols]> upper_bound, upper_bound, df[cols])
    df[cols]= np.where(df[cols]< lower_bound, lower_bound, df[cols])
```
